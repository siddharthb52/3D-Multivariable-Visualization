```python
import plotly.graph_objects as go
import numpy as np

# def f(x, y):
#   return x**2 + y**2

# Prompt user to set the ranges
print("Define the range (aka the minimum and maximum values for the X-Y plane of the graph\n")
user_x_min = input("Enter the minimum value of the x-axis: ")
x_min = float(user_x_min)
user_x_max = input("Enter the maximum value of the x-axis: ")
x_max = float(user_x_max)
user_y_min = input("Enter the minimum value of the y-axis: ")
y_min = float(user_y_min)
user_y_max = input("Enter the maximum value of the y-axis: ")
y_max = float(user_y_max)

x = np.linspace(x_min, x_max, 100)
y = np.linspace(y_min, y_max, 100)
X, Y = np.meshgrid(x, y)

# Prompt user to define variable Y based on X
user_expression = input("Enter the expression for Z in terms of X and Y: ")

# Use eval() to evaluate the user input as a mathematical input defined by an expression following Python syntax (eg. using ** for exponentiation rather than ^)
Z = eval(user_expression)

# Create a 3D surface plot
fig = go.Figure(data=[go.Surface(x=X, y=Y, z=Z, colorscale='magma')])

# Set axis labels
fig.update_layout(scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Probability Density'))

# Show the plot
fig.show()
```
